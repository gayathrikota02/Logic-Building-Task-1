# **Problem 1: Mobile Recharge Validation System**


```python
def validate(amount):
    if amount < 50:
        print("Amount must be at least ₹50")
        return False

    if amount != 199 and amount != 299 and amount != 399 and amount != 599:
        print("Invalid plan. Choose 199, 299, 399, or 599")
        return False

    return True


while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate(amount):
        print("Recharge Successful")
        break
    else:
        print("Try again\n")
```

# **Problem 2: Inventory Reorder Alert System**


```python
def check_stock(products):
    for name, quantity in products.items():
        if quantity < 15:
            print(name, "- Reorder Alert")
        else:
            print(name, "- Stock OK")


# Example dictionary
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 8
}

check_stock(inventory)
```

# **Problem 3: Student Result Processing System**


```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    # Determine result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example usage
student_marks = [60, 45, 70, 50, 55]
avg, res = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", res)
```

# **Problem 4: Cab Fare Estimator with Retry**

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12

    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra during peak hour
    if peak_hour.lower() == "yes":
        fare += fare * 0.25

    return fare


# Retry option using while loop
while True:
    distance = float(input("Enter distance (in km): "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator ")
        break
```

# **Problem 5: Employee Attendance Eligibility Checker**

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # Determine eligibility
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
percent, status = check_eligibility(attendance)

print("Attendance Percentage:", percent, "%")
print("Status:", status)
```

# **Problem 6: Password Strength Checker**

```python
def check_password_strength(password):
    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    # Loop over string
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


# Example usage
pwd = input("Enter your password: ")
result = check_password_strength(pwd)
print(result)
```
