# **Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_check(rainfall,required_level):
  average=sum(rainfall)/len(rainfall)
  print("Average Rainfall:",int(average))
  if average>=required_level:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")
rainfall=list(map(int,input().split()))
required_level=int(input())
rainfall_check(rainfall,required_level)

```

# **Problem 4: Social Media – Duplicate Account Detection**

```python
def Duplicate_Detection(data):
  unique=set(data)
  if len(unique)!=len(data):
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate ccounts Found :No")
data=input().split()
Duplicate_Detection(data)
```

# **Problem 10: Smart Electricity Bill Analyzer**

```python
def Bill_Analyzer(units):
  bill=0
  if 0<units<=100:
    bill=units*3
  elif 100<units<=200:
    bill=(100*3)+((units-100)*5)
  else:
    bill=(100*3)+(100*5)+((units-200)*7)
  print("Total Bill:",bill)

  if bill<500:
    print("Low Usage")
  elif 500<=bill<=1500:
    print("Moderate Usage")
  else:
    print("High Usage")
units=int(input())
Bill_Analyzer(units)

```
