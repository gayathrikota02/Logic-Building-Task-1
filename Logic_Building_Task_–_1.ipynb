```python
#1. User Login Check
username=input("Enter Your username:")
password=input("Enter Your password:")
if(username == "admin" and password == "1234"):
  print("Login Successful")
else:
  print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if(i>=50):
    pass_count=pass_count+1
  else:
    fail_count =fail_count+1
print("Total Students:",len(marks))
print("Count the total number of pass students:",pass_count)
print("Count the total number of fail students:",fail_count)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
new_list=[]
for i in names:
  new_list.append(i.strip().lower())
print(new_list)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  print(f"{i}, Length={len(i)}")
  if(len(i)>10):
    print("FLAG: Long Message")
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for i in logs:
  if(i=="ERROR"):
    error_count=error_count+1
print("Total Error Count:",error_count)
```
