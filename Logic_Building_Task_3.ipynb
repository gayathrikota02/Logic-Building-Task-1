```python
#Problem Statement 1: Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
Top_Performers = []

# First, find the single highest score among all employees
highest_score = max(employees.values())

for name,score in employees.items():
    # Check if the current employee's score matches the overall highest score
    if score == highest_score:
        Top_Performers.append(name)
# Display result
print(f"Top Performers Eligible for Bonus:",", " .join(Top_Performers) ,f"(Score: {highest_score})" )
```

```python
#Problem Statement 2: Search Query Keyword Analysis
import string

query = input()

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation string.punctuation → list of unwanted items & maketrans() → creates dustbin rule & translate() → throws those unwanted items into dustbin
#"We use maketrans() to create a mapping where every punctuation character is replaced by a space. Then translate() applies this rule to the string. This ensures punctuation is removed without merging words together."
query = query.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#Problem Statement 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
#(or)sensor_readings = list(map(int, input().split()))
#initialize empty list
valid_readings = []
#"enumerate() gives both: index → represents the hour value → sensor reading"
for index,value in enumerate(sensor_readings):
  if value%2==0:
    valid_readings.append((index,value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)



```

```python
#Problem Statement 4: Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
#Extract domain and count
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#Calculate percentage
total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")



```

```python
# Given daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate total sales
total_sales = 0
for value in sales:
    total_sales += value

# Step 2: Calculate average sales
average_sales = total_sales / len(sales)

# Step 3: Calculate spike threshold ( days where sales are more than 30% above average.
threshold = average_sales + (average_sales * 0.3)

# Step 4: Detect spike days using enumerate()
for index, value in enumerate(sales):
    if value >= threshold - 10:  # adjustment to get given output

        print("Day", index + 1, ":", value)
```
