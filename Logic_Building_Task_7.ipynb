# **Problem 1: Smart Light Controller**

```python
class Light:
  def __init__(self,name):
    self.name=name
    self.status="off"
  def switch_on(self):
    self.status="On"
  def switch_off(self):
    self.status="off"
  def show_status(self):
    print(f"{self.name} is {self.status}")

light_Name=input("Light Name:")
action=input("Action:")
light=Light(light_Name)
if action.upper()=="ON":
  light.switch_on()
elif action.lower()=="OFF":
  light.switch_off()
else:
  print("Invalid action")
  exit()
light.show_status()

```

# **Problem 2: Employee ID Card System**

```python
class Employee:
  def __init__(self,name,ID,dept):
    self.name=name
    self.ID=ID
    self.dept=dept
  def display_details(self):
    print("Employee ID Card")
    print(f"Name:{self.name}")
    print(f"ID:{self.ID}")
    print(f"Department:{self.dept}")

name=input("Name:")
id=input("ID:")
dept=input("Department:")
e=Employee(name,id,dept)
e.display_details()
```

# **Problem 3: Mobile Contact Record**

```python
class Mobile_Contact:
  def __init__(self,name,phonenumber):
    self.name=name
    self.phonenumber=phonenumber
  def display_details(self):
    print("Contact Saved")
    print(f"Name:{self.name}")
    print(f"Phone:{self.phonenumber}")

name=input("Name:")
phone=int(input("Phone:"))
m=Mobile_Contact(name,phone)
m.display_details()
```

# **Problem 4: Product Price Tag Generator**

```python
class Product_Price_Tag:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def display_details(self):
    print(f"Product:{self.name}")
    print(f"Price:₹{self.price}")
name=input("Product Name:")
price=int(input("Price:"))
p=Product_Price_Tag(name,price)
p.display_details()
```

# **Problem 5: Movie Rating Display System**

```python
class Movie_Rating_Display:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def display_details(self):
    print(f"Movie:{self.name}")
    print(f"Rating:{self.rating:.1f}/5")
name=input("Movie Name:")
rating=float(input("Rating:"))
m=Movie_Rating_Display(name,rating)
m.display_details()

```
