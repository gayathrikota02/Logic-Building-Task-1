# **Problem Statement 1:Smart Parking Lot Management System**

```python
def manage_parking_lot(capacity, logs):
    parked_vehicles = 0

    # Loop for counting entries and exits
    for log in logs:
        if log == "IN":
            parked_vehicles += 1
        elif log == "OUT":
            parked_vehicles -= 1

    # Conditional alert for parking status
    if parked_vehicles >= capacity:
        status = "Full"
    else:
        status = "Available"

    print(f"Currently Parked Vehicles: {parked_vehicles}")
    print(f"Parking Status: {status}")

# Sample Input Execution
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking_lot(50, vehicle_logs)
```

# **Problem Statement 2:Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    total_time = distance * 5

    # Add delay based on traffic level
    if traffic == "High":
        total_time += 15
    elif traffic == "Medium":
        total_time += 10
    else:
        total_time += 5

    # Add delay based on weather condition
    if weather == "Rainy":
        total_time += 10
    elif weather == "Storm":
        total_time += 20

    print("Estimated Delivery Time:", total_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

# **Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    # Calculate total booked seats from the list
    booked_count = len(booked_seats)

    # Percentage calculation
    occupancy_pct = (booked_count / total_seats) * 100

    # Determine if show is Housefull or Almost Full
    if occupancy_pct >= 100:
        status = "Housefull"
    elif occupancy_pct >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Occupancy: {int(occupancy_pct)}%")
    print(f"Show Status: {status}")

# Sample Input Execution
# Creating a list of 150 entries to match the sample input requirement
booked_seats_list = [1] * 150
analyze_seat_occupancy(200, booked_seats_list)
```

# **Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage
    for value in cpu_readings:
        total += value

    average_load = total / len(cpu_readings)

    print("Average CPU Load:", int(average_load), "%")

    if average_load < 50:
        print("Server Status: Normal")
    elif average_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

# **Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def classroom_resources(usage_dict):

    overused_items = []

    for item in usage_dict:
        if usage_dict[item] > 8:   # more than 8 hours considered overuse
            overused_items.append(item)

    if len(overused_items) > 0:
        print("Overused Resources:", ", ".join(overused_items))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


classroom_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

# **Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_registration(capacity, total_registrations):

    if total_registrations <= capacity:

        confirmed = total_registrations
        waitlist = 0
        status = "Open"

    else:

        confirmed = capacity
        waitlist = total_registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


manage_registration(100, 105)
```
